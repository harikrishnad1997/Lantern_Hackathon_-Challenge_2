```python
import pandas as pd
import h2o
h2o.init()
```

```python
df = h2o.import_file(path="Rapamycin_final.csv")
df.head()
```

```python
df.shape
```

```python
y = "Rapamycin_response"
x = df.columns
x.remove(y)
len(x)
```

```python
df[y] = df[y].asfactor()
```

```python
for col in x:
    if 'mut' in col:
        df[col] = df[col].asfactor()
```

```python
df["type"] = df["type"].asfactor()
```

```python
from h2o.automl import H2OAutoML
aml = H2OAutoML(max_models=100,balance_classes=True, seed=72854874
                # ,include_algos = ["DRF","StackedEnsemble" , "XGBoost"]
               )
aml.train(x=x,y=y, training_frame=df)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
best_model = aml.get_best_model()
print(best_model)
```

```python
aml.leader
```

```python
h2o.save_model(model=best_model, path="/root/capsule/code/Rapamycin", force=True)
```
