```python
import pandas as pd
raw = pd.read_csv("/root/capsule/data/challenge2_files/fulvestrantTrain.tsv",sep='\t')
raw.head()
```

```python
import h2o
h2o.init()
```

```python
columns = pd.read_csv("/root/capsule/code/final_col_fulvestrant.csv",index_col=0)
col = list(set(columns["features"]))
len(col)
```

```python
df = h2o.H2OFrame(raw[col])
df.describe()
```

```python
y = "Fulvestrant_response"
x = df.columns
x.remove(y)
len(x)
```

```python
df[y] = df[y].asfactor()
```

```python
for col in x:
    if 'mut' in col:
        df[col] = df[col].asfactor()
```

```python
df["type"] = df["type"].asfactor()
```

```python
c=0
for col in df.types:
    if 'rna' in col:
        print(df[col].types)
        c+=1
        if c>10:
            break
```

```python
df.head()
```

```python
from h2o.automl import H2OAutoML
```

```python
aml = H2OAutoML(max_models=100,balance_classes=True, seed=72854874
                # ,include_algos = ["DRF","StackedEnsemble" , "XGBoost"]
               )
aml.train(x=x,y=y, training_frame=df)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
best_model = aml.get_best_model()
print(best_model)
```
