```python
from h2o.automl import H2OAutoML
```

```python
import pandas as pd
df = pd.read_csv("/root/capsule/data/challenge2_files/fulvestrantTrain.tsv",sep='\t')
df.head()
```

```python
import h2o
h2o.init()
```

```python
df = h2o.H2OFrame(df)
```

```python
df.head()
```

```python
columns = pd.read_csv("/root/capsule/code/final_col_fulvestrant.csv",index_col=0)
col = list(set(columns["features"]))
col[0:5]
```

```python
"Fulvestrant_response" in col
```

```python
y = "Fulvestrant_response"
x = [element for element in col if element != y]
x[0:5]
```

```python
"Fulvestrant_response" in x
```

```python
df[y] = df[y].asfactor()
```

```python
df[y]
```

```python
for col in x:
    if col.startswith('mut'):
        df[col] = df[col].asfactor()
```

```python
df["type"] = df["type"].asfactor()
```

```python
train, test = df.split_frame(ratios=[0.8], seed = 90756174)
```

```python
aml = H2OAutoML(max_models=10,balance_classes=True, seed=72854874, exclude_algos = ["DeepLearning"])
aml.train(y=y, training_frame=train,validation_frame = test)
```
